```python
import pymysql
from sqlalchemy import create_engine
import pandas as pd
import getpass  # to get the password without showing the input
password = getpass.getpass()
```

```python
connection_string = 'mysql+pymysql://root:' + password + '@localhost/sakila'
engine = create_engine(connection_string)

rentals_from_may = pd.read_sql_query('select customer_id, count(rental_id) as may from rentals_may group by customer_id', engine)
rentals_from_may
```

```python
rentals_from_june = pd.read_sql_query ('select customer_id, count(rental_id) as june from rentals_june group by customer_id', engine)
rentals_from_june
```

```python
new = rentals_from_may.merge(rentals_from_june, on = 'customer_id')
new
```

```python
may = 2
june = 6
if may > june:
    print('May is higher')
elif may < june:
    print('June is higher')
else:
    print('Equal')
```

```python
def sanja(dataframe):
    if dataframe['may'] > dataframe['june']:
        return 'May is higher'
    elif dataframe['may'] < dataframe['june']:
        return 'June is higher'
    else:
        return 'Equal'
new['compare'] = new.apply(sanja, axis =1)
new
```

```python
import numpy as np
conditions = [
    (new['may'] > new['june']), (new['may'] < new['june']), (new['may']== new['june'])]

choices = ['May', 'June', 'Equal']

new['compare'] = np.select(conditions, choices, default=np.nan)
new
```
